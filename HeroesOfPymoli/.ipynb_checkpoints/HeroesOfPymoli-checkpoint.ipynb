```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
purchase_data
```

## Player Count

* Display the total number of players


```python
total_players = len(purchase_data["SN"].value_counts())
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Create Variables to hold the values of Summary Calculations; Apply appropriate formats
total_unique_items = len(purchase_data["Item Name"].value_counts())
total_unique_items

average_price = purchase_data["Price"].mean()
str_avg_price = '$%.2f' % (average_price)


total_purchase = len(purchase_data["Purchase ID"].value_counts())

total_revenue = purchase_data["Price"].sum()
str_total_rev = '$%.2f' % (total_revenue)

#Create dictionary and summary dataframe
summary_dict = [{"Number of Unique Items":total_unique_items,"Average Price": str_avg_price ,"Number of Purchases": total_purchase,"Total Revenue":str_total_rev} ]
summary_df = pd.DataFrame(summary_dict)

summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Create dataframe per Gender from purchase_data
male = purchase_data.loc[purchase_data["Gender"] == "Male"]
female = purchase_data.loc[purchase_data["Gender"] == "Female"]
other = purchase_data.loc[(purchase_data["Gender"] != "Female") & (purchase_data["Gender"] != "Male")]

# Calculate Percentages and totals per Gender
male_total = len(male["SN"].unique())
male_per = "{:.2%}".format( (male_total/total_players))

female_total = len(female["SN"].unique())
female_per = "{:.2%}".format( female_total/total_players)

other_total = len(other["SN"].unique())
other_per = "{:.2%}".format( other_total/total_players)

# Create Summary dataframe with calculated percentages and totals
gender_df =pd.DataFrame([{"Gender" : "Male", "Total Count" : male_total, "Percentage of Players" : male_per},
              {"Gender" : "Female", "Total Count" : female_total, "Percentage of Players" :female_per},
              {"Gender":"Other/Non-Discolosed","Total Count" : other_total, "Percentage of Players" : other_per}])
gender_sum = gender_df.set_index("Gender")
gender_sum
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Create dataframe with set columns
purchase_gender_df = pd.DataFrame(columns = ["Gender","Purchase Count","Average Purchase Price","Total Purchase Value","Avg Total Purchase per Person"])
gender = ["Female","Male","Other/Non-Disclosed"]

# Create lists for dataframe series per designated columns; pull data from gender dataframes from prev: Gender Demographics
purch_count = [female["Purchase ID"].count(),male["Purchase ID"].count(),other["Purchase ID"].count()]
avg_purch = ['$%.2f' % female["Price"].mean(),'$%.2f' % male["Price"].mean(),'$%.2f' % other["Price"].mean()]
total_value = ['$%.2f' % female["Price"].sum(),'$%.2f' % male["Price"].sum(),'$%.2f' % other["Price"].sum()]
avg_purch_per = ['$%.2f' % (female["Price"].sum()/female_total),'$%.2f' % (male["Price"].sum()/male_total),'$%.2f' % (other["Price"].sum()/other_total)]

# Assign series values from lists
purchase_gender_df["Gender"] = gender
purchase_gender_df["Purchase Count"] = purch_count
purchase_gender_df["Average Purchase Price"] = avg_purch
purchase_gender_df["Total Purchase Value"] = total_value
purchase_gender_df["Avg Total Purchase per Person"] = avg_purch_per

# Set Index for dataframe
purchase_gender_df = purchase_gender_df.set_index("Gender")
purchase_gender_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create bins and labels for Age Groups
bins = [0,9,14,19,24,29,34,39,100]
age_grps = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

# Remove duplicates of SN to get count of existing players and Create column Age Group as assigned by bins
individual_df = purchase_data.drop_duplicates(subset="SN")
individual_df["Age Group"] = pd.cut(individual_df["Age"],bins,labels=age_grps, include_lowest=True)
individual_df.head()

# Create dataframe grouped by Age Group
age_dem_df = pd.DataFrame({'Total Count': individual_df.groupby(["Age Group"]).size()})

# Calculate Percentage of Players using GroupBy cumulative functions; Apply appropriate formatting
age_count= age_dem_df["Total Count"]
age_per = []

for i in range(len(age_count)):
    age_per.append("{:.2%}".format(age_count[i]/total_players))

# Add Percentage of Players column to dataframe
age_dem_df["Percentage of Players"] = age_per
age_dem_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Bin purchase_data dataframe by age ----> already done in last prompt: Age Demographics
# Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc.
age_price = individual_df[["Price","Age Group"]]
age_purch_group = age_price.groupby("Age Group")

age_group_avg = age_purch_group.mean()
age_group_sum = age_purch_group.sum()
age_group_per = []

# Assign groupby objects to series to add to summary dataframe
age_group_avg_series = age_group_avg["Price"]
age_group_sum_series = age_group_sum["Price"]

for j in range(0,len(age_count)):
    age_group_per.append("${:.2f}".format(age_group_sum_series[j]/age_count[j]))
    
# Create summary table and apply appropriate formats   
age_summary = pd.DataFrame({'Purchase Count': individual_df.groupby(["Age Group"]).size()})
age_summary["Average Purchase Price"] = age_group_avg_series.map('${:,.2f}'.format)
age_summary["Total Purchase Value"] = age_group_sum_series.map('${:,.2f}'.format)
age_summary["Avg Total Purchase per Person"] = age_group_per

age_summary
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Set dataframe to hold only Price and SN information then group by SN
sn_price = purchase_data[["Price","SN"]]
group_sn = sn_price.groupby(purchase_data["SN"])

# Calculate the appropriate calculations for groups
group_sn_avg = group_sn.mean()
group_sn_sum = group_sn.sum()

# Assign groupby objects to series to add to summary dataframe
group_sn_avg_series = group_sn_avg["Price"]
group_sn_sum_series = group_sn_sum["Price"]

# Create summary table and apply appropriate formats; Sort by Descending Purchase Count
sn_summary = pd.DataFrame({'Purchase Count': purchase_data.groupby(["SN"]).size()})
sn_summary["Average Purchase Price"] = group_sn_avg_series.map('${:,.2f}'.format)
sn_summary["Total Purchase Value"] = group_sn_sum_series.map('${:,.2f}'.format)
sort_sn_summary = sn_summary.sort_values(["Purchase Count"], ascending=False)
sort_sn_summary.head() 
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Set dataframe to hold only Price and SN information then group by SN
item_df = purchase_data[["Item ID","Item Name","Price"]]
group_item = item_df.groupby(["Item ID","Item Name"])

# Calculate the appropriate calculations for groups
group_id_sum = group_item.sum()

# Assign groupby objects to series to add to summary dataframe
group_id_sum_series = group_id_sum["Price"]

# Create summary table and apply appropriate formats
sn_summary = pd.DataFrame({'Purchase Count': purchase_data.groupby(["Item ID","Item Name"]).size()})
sn_summary["Total Purchase Value"] = group_id_sum_series.map('${:,.2f}'.format)

# Create an Items data frame grouped by same Indexes Item ID and Item Name to add Items Price to merge_sn dataframe
item = purchase_data.drop_duplicates(subset="Item ID")
item_price = item.groupby(["Item ID","Item Name"])
items=item_price.sum()

merge_sn = sn_summary.join(items)
merge_sn["Price"] = items["Price"].map('${:,.2f}'.format)

# Sort sn_summary dataframe by descending Purchase Count; Reorganize Columns and Column names to match example
sort_sn_summary = merge_sn.sort_values(["Purchase Count"], ascending=False)
sort_summary = sort_sn_summary.drop(columns=['Purchase ID','Age'])
summary = sort_summary[["Purchase Count","Price","Total Purchase Value"]]
summary = summary.rename(columns = {"Price":"Item Price"})
summary.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
# Create summary table and apply appropriate formats
profit_summary = pd.DataFrame({'Purchase Count': purchase_data.groupby(["Item ID","Item Name"]).size()})
profit_summary["Total Purchase Value"] = group_id_sum_series

# Create an Items data frame grouped by same Indexes Item ID and Item Name to add Items Price to merge_sn dataframe
profit_item = purchase_data.drop_duplicates(subset="Item ID")
profit_item_price = profit_item.groupby(["Item ID","Item Name"])
profit_items=profit_item_price.sum()


profit_merge_sn = profit_summary.join(profit_items)
profit_merge_sn["Price"] = profit_items["Price"].map('${:,.2f}'.format)


sort_sn_profit = profit_merge_sn .sort_values(["Total Purchase Value"], ascending=False)
sort_sn_profit["Total Purchase Value"] = sort_sn_profit["Total Purchase Value"].map('${:,.2f}'.format)
profit_summary = sort_sn_profit.drop(columns=['Purchase ID','Age'])
p_summary = profit_summary[["Purchase Count","Price","Total Purchase Value"]]
p_summary = p_summary.rename(columns = {"Price":"Item Price"})
p_summary.head()
```
